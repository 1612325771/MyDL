```python
with open('animal.csv', mode='w', newline='') as f:
    writer = csv.writer(f)
    for image in image_dir:
        class_name = image.split(os.sep)[-2]
        label = species_to_idx[class_name]
        writer.writerow([image, label])
```

```python

trans = transforms.Compose([
    transforms.Resize([224,224]),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[.5, .5, .5], std=[.5, .5, .5])
])
```

```python
class makedataset(Dataset):
    def __init__(self, csv_filename, mode,transform):
        super(makedataset, self).__init__()
        self.csv_filename = csv_filename
        self.mode = mode
        self.transforms = transform
        self.image, self.label = self.load_csv()
        if mode == 'train':
            self.image = self.image[:int(0.8*len(self.image))]
            self.label = self.label[:int(0.8*len(self.label))]
        elif mode == 'val':
            self.image = self.image[int(0.8*len(self.image)):]
            self.label = self.label[int(0.8*len(self.label)):]
    def load_csv(self):
        image,label = [], []
        with open(self.csv_filename) as f:
            reader = csv.reader(f)
            for row in reader:
                i, l = row
                image.append(i)
                label.append(int(l))
        return image,label
    def __len__(self):
        return len(self.image)
    def __getitem__(self, index):
        img = self.image[index]
        label_tensor = torch.tensor(self.label[index])
        pil_img = Image.open(img)
        pil_img = pil_img.convert("RGB")
        data = self.transforms(pil_img)
        return data, label_tensor
```

```python
train_db = makedataset('/home/bear/dl/animal.csv', 'train', trans)
train_db[2]
```
